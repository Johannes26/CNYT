##  1. Suma de numeros complejos

La funcion Sumar_r declara las variables para las cuales lee 2 numeros complejos y determina la suma entre ellos.

```python
def Sumar_Comp(C1,C2):
    #declaraci√≥n de variables para la lectura de numeros complejos, en el cual lee 2 numeros separados por un espacio;
    #El primero es la parte real y el segundo la parte imaginaria
    Real=C1[0]+C2[0]
    Comp=C1[1]+C2[1]
    
    return((Real,Comp))
```

## 2. Multiplicacion de numeros complejos

```python
def Multi_Comp(C1,C2): 
    Real=(C1[0]*C2[0])-(C1[1]*C2[1])
    Comp=(C1[0]*C2[1])+(C1[1]*C2[0])
    
    return((Real,Comp))
```

## 4. Division

```python
def Division_Comp(C1,C2):
    if C2==(0,0): 
        print("No se puede dividir entre 0")
    else:
        #Conjugado
        Conj=(C2[0],-C2[1])

        #Multiplicacion Denominador
        Denom=Multi_Comp(Conj,C2)

        #Multiplicacion Numerador
        Num=Multi_Comp(C1,Conj)

        res=(Num[0]/Denom[0],Num[1]/Denom[0])
        return(res)
```

```python
def Mat_Trans(M1):
    M2=[]
    for i in range(len(M1[0])):
        fila=[]
        for j in range(len(M1)):
            fila.append(M1[j][i])
        M2.append(fila)
    return M2
```

## 24. Producto tensor

```python
def tensor(M1,M2):
    matriz=[]
    m=len(M1)
    n=len(M1[0])
    p=len(M2)
    q=len(M2[0])
    for i in range(m*p):
        fila=[]
        for j in range(n*q):
            fila.append(Multi_Comp(M1[i//p][j//q],M2[i%p][j%q]))
        matriz.append(fila)
    return matriz
```

# 24. Producto tensor vectores

```python
def tensorvectores(V1,V2):
    fila=[]
    for j in range(len(V1)*len(V2)):
        fila.append(Multi_Comp(V1[j//len(V2)],V2[j%len(V2)]))
    return fila
```

```python
def rendijas(dinamica,num_rendijas,num_blancos_pared,vector_prob):
    dimension=1+(2*num_rendijas)+(num_blancos_pared*(num_rendijas+1))
    estado_ini=[(1,0)]+([(0,0)]*(dimension-1))
    if(dinamica==1):
        prob_blancos=1/(((num_blancos_pared*2)+1)**(1/2))
        vec=[]
        for i in range(len(estado_ini)):
            vec.append(Multi_Comp(estado_ini[i],Conjugado_Comp(estado_ini[i])))
        estado_ini=vec
    else:
        prob_blancos=1/((num_blancos_pared*2)+1)
    if (vector_prob==[]):
        vector_prob=[(prob_blancos,0)]*dimension
    prob_ren=1/num_rendijas
    contacto=(2*num_blancos_pared)+1
    M=[]
    pos=num_rendijas
    for col in range(dimension):
        columna=[]
        cont=0
        for fil in range(dimension):
            if(col==0):
                if(fil>0 and fil<=num_rendijas and dinamica==0):
                    columna.append((prob_ren,0))
                elif(fil>0 and fil<=num_rendijas and dinamica==1):
                    columna.append(((1/(prob_ren**(1/2))),0))
                else:
                    columna.append((0,0))
            elif(col>0 and col<=num_rendijas):
                if(fil>pos and cont!=contacto):
                    columna.append(vector_prob[cont])
                    pos+=1
                    cont+=1
                else:
                    columna.append((0,0))
            else:
                if(col==fil):
                    columna.append((1,0))
                else:
                    columna.append((0,0))
        M.append(columna)
        if(col!=0):
            pos-=num_blancos_pared
    M=Mat_Trans(M)
    M2=prod_Mat(M,M)
    if(dinamica==1):
        matrizModulos=[]
        for x in range(len(M2)):
            fila=[]
            for y in range(len(M2[0])):
                fila.append(Multi_Comp(M2[x][y],Conjugado_Comp(M2[x][y])))
            matrizModulos.append(fila)
        M2=matrizModulos
    
    vectorFinal=accion(M2,estado_ini)
    labels=[]
    estado=[]
    for x in range(len(vectorFinal)):
        if x==0:
            labels.append('P.'+str(x))
        elif x>0 and x<=num_rendijas:
            labels.append('R.'+str(x))
        else:
            labels.append('B.'+str(x))
        estado.append(vectorFinal[x][0])
    index = np.arange(len(labels))
    plt.bar(index, estado)
    plt.xlabel('Estado')
    plt.ylabel('Valor')
    plt.xticks(index, labels, rotation=90)
    plt.title('Evoluci√≥n din√°mica del sistema')
    plt.show()
    return vectorFinal
```

```python
def SistemaCuantico(V):
    suma=(0,0)
    for x in V:
        suma=Sumar_Comp(suma,Multi_Comp(x,Conjugado_Comp(x)))
    vector=[]
    for x in V:
        vector.append(((Multi_Comp(x,Conjugado_Comp(x))[0]/suma[0]),0))
    return vector
```

```python
def VectorNormalizado(V):
    suma=(0,0)
    for x in V:
        suma=Sumar_Comp(suma,Multi_Comp(x,Conjugado_Comp(x)))
    suma=(suma[0]**(1/2),0)
    vectornom=[]
    for x in V:
        vectornom.append((x[0]/suma[0],x[1]/suma[0]))
    return vectornom
```
